```python
#Drop the columns
df.drop(["ID","UniqueCarrier","TailNum","Origin","Dest","CancellationCode"],axis =1,inplace =True)
```

```python
X = df.drop("Cancelled",axis =1)
y = df["Cancelled"]
```

```python
#Splitting the data into train and test split to build Naive Bayes Model
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state=42)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train,y_train)
```

```python
gnb.score(X_train,y_train)
```

```python
y_train_pred = gnb.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_train,y_train_pred))
```

```python
Confusion_MatrixData = confusion_matrix(y_train,y_train_pred)

#Calculating Accuracy,Percision,Recall and F1-Score
Accuracy = (Confusion_MatrixData[1][0] + Confusion_MatrixData[0][0])/(Confusion_MatrixData[0][0] + Confusion_MatrixData[0][1] + Confusion_MatrixData[1][0]+ Confusion_MatrixData[1][1])
Senstivity = (Confusion_MatrixData[1][1]) /  (Confusion_MatrixData[1][1] + Confusion_MatrixData[1][0])
Precision = (Confusion_MatrixData[1][1]) / (Confusion_MatrixData[1][1] + (Confusion_MatrixData[0][1])) 
F1Score = (2 * Senstivity * Precision)/(Senstivity + Precision)
```

```python
Performance_Metrics = pd.DataFrame({"Model": "Naive Bayes[Train]","Accuracy": [Accuracy],"Precision":[Precision],"Recall":[Senstivity],"F1Score":[F1Score]})
Performance_Metrics
```

```python
#Checking test data performance
y_pred_test = gnb.predict(X_test)
y_pred_test
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
Confusion_MatrixData = confusion_matrix(y_test,y_pred_test)

#Calculating Accuracy,Percision,Recall and F1-Score
Accuracy = (Confusion_MatrixData[1][0] + Confusion_MatrixData[0][0])/(Confusion_MatrixData[0][0] + Confusion_MatrixData[0][1] + Confusion_MatrixData[1][0]+ Confusion_MatrixData[1][1])
Senstivity = (Confusion_MatrixData[1][1]) /  (Confusion_MatrixData[1][1] + Confusion_MatrixData[1][0])
Precision = (Confusion_MatrixData[1][1]) / (Confusion_MatrixData[1][1] + (Confusion_MatrixData[0][1])) 
F1Score = (2 * Senstivity * Precision)/(Senstivity + Precision)
```

```python
Performance_Metrics_test = pd.DataFrame({"Model": "Naive Bayes[Test]","Accuracy": [Accuracy],"Precision":[Precision],"Recall":[Senstivity],"F1Score":[F1Score]})
Performance_Metrics_test
```

```python
pd.concat([Performance_Metrics,Performance_Metrics_test],axis =0)
```
