```python
import pandas as pd
```

```python
data=pd.read_csv(r'C:\Users\ravig\Downloads\Cleaned Data.csv')
```

```python
# Create a histogram of the TaxiIn field
import matplotlib.pyplot as plt
data['TaxiIn'].plot.hist(bins=50, edgecolor='black')

# Set the plot title and axis labels
plt.title('Distribution of Taxi-In Time')
plt.xlabel('Taxi-In Time (minutes)')
plt.ylabel('Number of Flights')

# Display the plot
plt.show()
```

```python
# Create a scatter plot of TaxiIn times by Arrival Delay
plt.scatter(data['ArrDelay'], data['TaxiIn'])

# Set the plot title and axis labels
plt.title('Taxi-In Time vs. Arrival Delay')
plt.xlabel('Arrival Delay (minutes)')
plt.ylabel('Taxi-In Time (minutes)')

# Display the plot
plt.show()
```

```python
data.boxplot(column='TaxiIn', by='UniqueCarrier')

# Set the plot title and axis labels
plt.title('Taxi-In Time by Unique Carrier')
plt.xlabel('Unique Carrier')
plt.ylabel('Taxi-In Time (minutes)')

# Rotate the x-axis labels for readability
plt.xticks(rotation=90)

# Display the plot
plt.show()
```
