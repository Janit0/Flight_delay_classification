# Replacing zero values with null values

```python
df['DepDelay'].replace(0, np.nan, inplace=True)
```

```python
df['DepDelay'].isnull().sum()
```

# formula for calculating the departure delay

```python
formula= df.ArrDelay + abs(df.ActualElapsedTime-df.CRSElapsedTime)
```

# filling up the null values with the formulated values

```python
df.DepDelay.fillna((formula), inplace=True)
```
